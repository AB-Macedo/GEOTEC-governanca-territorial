# <span style="color:#000000">4. Cálculo do Indicador: Dinâmica do CAR </span>
<hr style="border:1px solid #000000;">

<p>
Nesta etapa, iremos analisar a evolução temporal de um imóvel do Cadastro Ambiental Rural (CAR) ao longo de diferentes versões da base de dados.
</p>

<p>
O indicador está estruturado em duas dimensões complementares: <b>ocorrência</b> e <b>intensidade</b>.
</p>

### Ocorrência

A ocorrência avalia se houve alteração na geometria do imóvel entre versões consecutivas:

<ul>
  <li><b>(Métrica A) — Ocorrência de alteração geométrica:</b> identifica se houve alteração na geometria, a partir da comparação de equivalência espacial (<code>ST_Equals</code>).</li>
</ul>

### Intensidade

A intensidade quantifica a extensão das alterações observadas, considerando:

<ul>
  <li><b>(Métrica B) — Número de alterações:</b> contabiliza quantas mudanças ocorreram ao longo das versões;</li>
  <li><b>(Métrica C) — Mudança de porte:</b> verifica se o imóvel mudou de classificação de tamanho (pequeno, médio ou grande) entre versões;</li>
  <li><b>(Métrica D) — Variação de área:</b> calcula a diferença percentual entre as áreas do imóvel em versões consecutivas;</li>
  <li><b>(Métrica E) — Percentual de área sobreposta:</b> calcula o percentual de sobreposição entre a geometria da versão anterior e da versão posterior.</li>
</ul>

<br>

###  4.2.1: Selecionar e visualizar o CAR de Interesse

#### Seleciona e visualiza as informações do CAR

```python
# ============================================================================
# Consulta SQL para selecionar o imóvel em todas as versões
# ============================================================================

queries = []

for i, tabela in enumerate(CAR_TABLES, start=1):
    data_ref = CAR_VERSOES[tabela]   
    q = f"""
    (
        SELECT DISTINCT ON ({CAR_ID})
            {i} AS versao,
            DATE '{data_ref}' AS data_ref,
            {CAR_ID},
            split_part({CAR_ID}, '-', 2) AS cod_mun,
            municipio,
            {CAR_GEOM} AS geom
        FROM {CAR_SCHEMA}.{tabela}
        WHERE {CAR_ID} = '{COD_IMOVEL}'
          AND {CAR_STATUS} <> 'CA'
          AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
        ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
    )
    """
    queries.append(q)

query_car_versoes = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
)
SELECT *
FROM car_series
ORDER BY versao;
"""

# ============================================================================
# Exibição da query
# ============================================================================

html_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em; margin-bottom: 15px;">
<b>Consulta SQL - Selecao do imovel em todas as versoes (UNION ALL):</b><br>
<code>
{query_car_versoes}
</code>
</div>
"""

display(HTML(html_query))
```

#### Executa a consulta

```python
# ============================================================================
# Executa a consulta e carrega os resultados
# ============================================================================

df_car_versoes = pd.read_sql(text(query_car_versoes), con=engine)

# ============================================================================
# Formata a coluna geom para mostrar apenas os primeiros caracteres
# ============================================================================

def formatar_geom(geom_str):
    if isinstance(geom_str, str) and len(geom_str) > 50:
        resumo = geom_str[:50] + "..."
        return f'<details><summary style="cursor:pointer; color:#2196f3;">{resumo}</summary><code style="word-break:break-all;">{geom_str}</code></details>'
    return geom_str

df_car_versoes_display = df_car_versoes.copy()
if 'geom' in df_car_versoes_display.columns:
    df_car_versoes_display['geom'] = df_car_versoes_display['geom'].apply(formatar_geom)

# ============================================================================
# Exibe o resultado
# ============================================================================

html_resultado = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin-bottom: 15px;">
<b>Imovel selecionado - Todas as versoes</b><br>
<br>
{df_car_versoes_display.to_html(escape=False, index=False)}
</div>
"""

display(HTML(html_resultado))
```

#### Visualização das geometrias dos CARs


```python
from IPython.display import HTML, display
import folium
import gc
import random
from shapely import wkb
import binascii

# ============================================================================
# Visualização da geometria do CAR em todas as versões (UNION ALL)
# ============================================================================

# Verifica se os dados foram carregados
if 'df_car_versoes' not in locals() or df_car_versoes.empty:
    print("X Nenhum imóvel carregado. Execute a célula de carregamento primeiro.")
else:
    # -------------------------------------------------------------------------
    # Define cores para cada versão (mais visíveis)
    # -------------------------------------------------------------------------
    versoes_unicas = sorted(df_car_versoes['versao'].unique())
    
    # Cores mais visíveis (bordas escuras, preenchimento claro)
    # Quanto maior a versão, mais escura a borda (para dar noção de tempo)
    def get_cor_versao(versao_num, total_versoes):
        # Borda: do cinza claro ao cinza escuro
        intensidade = 0.3 + (versao_num - 1) * 0.5 / max(1, total_versoes - 1)
        intensidade = min(intensidade, 0.8)
        cor_borda = f"rgb({int(128 + intensidade * 127)}, {int(128 + intensidade * 127)}, {int(128 + intensidade * 127)})"
        # Preenchimento: sempre branco com opacidade
        return cor_borda
    
    # -------------------------------------------------------------------------
    # Função para converter WKB para geometria
    # -------------------------------------------------------------------------
    def wkb_to_geom(wkb_str):
        if isinstance(wkb_str, str):
            geom_bytes = binascii.unhexlify(wkb_str)
            return wkb.loads(geom_bytes)
        return wkb_str
    
    # -------------------------------------------------------------------------
    # Prepara o mapa base (usa a primeira versão para centralizar)
    # -------------------------------------------------------------------------
    primeira_linha = df_car_versoes.iloc[0]
    primeira_geom = wkb_to_geom(primeira_linha['geom'])
    
    bounds = primeira_geom.bounds
    map_center = [(bounds[1] + bounds[3]) / 2, (bounds[0] + bounds[2]) / 2]
    
    m = folium.Map(
        location=map_center,
        zoom_start=10,
        control_scale=True,
        tiles='Esri.WorldImagery',
        width='80%',
        height=500
    )
    m.fit_bounds([[bounds[1], bounds[0]], [bounds[3], bounds[2]]])
    
    # -------------------------------------------------------------------------
    # Adiciona cada versão ao mapa (da mais antiga para a mais recente)
    # -------------------------------------------------------------------------
    total_versoes = len(versoes_unicas)
    
    for idx, row in df_car_versoes.iterrows():
        versao_num = row['versao']
        data_ref = row['data_ref']
        cod_imovel = row['cod_imovel']
        municipio = row['municipio']
        cod_estado = row.get('cod_estado', '')
        
        # Converte a geometria
        geom = wkb_to_geom(row['geom'])
        
        # Prepara geometria para o Folium
        gdf_temp = gpd.GeoDataFrame({'geometry': [geom]}, crs=4674)
        gdf_json_safe = gdf_temp.copy()
        
        # Define a cor da borda baseada na versão
        cor_borda = get_cor_versao(versao_num, total_versoes)
        
        tooltip_text = f"""
        <div style="font-family: Arial, sans-serif;">
            <b>CAR - Versao {versao_num} ({data_ref})</b><br>
            <b>Codigo:</b> {cod_imovel}<br>
            <b>Municipio:</b> {municipio} {f'/ {cod_estado}' if cod_estado else ''}
        </div>
        """
        
        folium.GeoJson(
            gdf_json_safe,
            name=f"Versao {versao_num} ({data_ref})",
            style_function=lambda x, cor=cor_borda: {
                'color': cor,
                'weight': 2.5,
                'fillColor': 'white',
                'fillOpacity': 0.35
            },
            tooltip=folium.Tooltip(tooltip_text)
        ).add_to(m)
    
    folium.LayerControl().add_to(m)
    display(m)
    
    # Libera memória
    del m
    gc.collect()
```

### 4.2.2: Identificar se houve alteração de geometria

Nesta etapa, avaliamos se a geometria de um imóvel foi modificada ao longo do tempo, comparando versões consecutivas da base do CAR.

#### Cria tabela auxiliar para comparar as geometrias


Após organizar os dados em sequência temporal (etapa anterior), utilizamos a função `LAG` do SQL que permite acessar valores de linhas anteriores dentro de uma mesma partição de dados.

Nesta etapa, `LAG` é aplicado sobre a ordenação da variável `versao`, permitindo recuperar, para cada registro, os valores da versão imediatamente anterior:

- `LAG(geom)` retorna a geometria da versão anterior (`geom_anterior`);
- `LAG(versao)` retorna o identificador da versão anterior (`versao_anterior`).

A estrutura utilizada é:

`LAG(coluna) OVER (ORDER BY versao)`

Dessa forma, cada linha passa a conter não apenas suas próprias informações, mas também os dados da versão anterior, o que permite realizar comparações diretas entre versões consecutivas.

Essa abordagem é fundamental para identificar mudanças ao longo do tempo, como alterações na geometria do imóvel.

```python
# ============================================================================
# Query para criar tabela de comparação entre as versões
# ============================================================================

queries = []

for i, tabela in enumerate(CAR_TABLES, start=1):
    data_ref = CAR_VERSOES[tabela]
    
    q = f"""
    (
        SELECT DISTINCT ON ({CAR_ID})
            {i} AS versao,
            DATE '{data_ref}' AS data_ref,
            {CAR_ID},
            split_part({CAR_ID}, '-', 2) AS cod_mun,
            municipio,
            {CAR_GEOM} AS geom
        FROM {CAR_SCHEMA}.{tabela}
        WHERE {CAR_ID} = '{COD_IMOVEL}'
          AND {CAR_STATUS} <> 'CA'
          AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
        ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
    )
    """
    queries.append(q)

query_tab_paracomparar = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
)
SELECT
    *,
    LAG(geom) OVER (ORDER BY versao) AS geom_anterior,
    LAG(versao) OVER (ORDER BY versao) AS versao_anterior
FROM car_series;
"""

# ============================================================================
# Exibição da query
# ============================================================================

html_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em; margin-bottom: 15px;">
<b>Consulta SQL - Cria tabela auxiliar para comparar as geometrias:</b><br>
<code>
{query_tab_paracomparar}
</code>
</div>
"""

display(HTML(html_query))
```

#### Cálculo para verificar alteração de geometria entre versões

Para verificar se a geometria do imóvel foi alterada utilizamos a função espacial `ST_Equals`,esta função retorna TRUE quando duas geometrias são topologicamente iguais.
Isso significa que:
- As geometrias possuem a mesma dimensão (ex: ambas são polígonos)
- Seus conjuntos de pontos ocupam exatamente o mesmo espaço

Diferente de uma comparação direta, a igualdade topológica não depende da forma como a geometria é representada, mas sim do espaço que ela ocupa. Assim, mesmo com variações na ordem dos vértices ou na estrutura interna, o resultado pode ser TRUE, desde que a área seja equivalente. Por outro lado, qualquer alteração na forma, posição ou área resulta em FALSE. Mesmo pequenas mudanças na geometria (como ajustes de limite) fazem com que o resultado seja **FALSE**, ainda que a diferença não seja significativa do ponto de vista territorial. 

Assim, essa abordagem permite identificar de forma objetiva se houve ou não alteração geométrica entre versões, embora não capture a magnitude da mudança.


```python
# ============================================================================
# Query para verificar alteração de geometria entre versões (ST_Equals)
# ============================================================================

queries = []

for i, tabela in enumerate(CAR_TABLES, start=1):
    data_ref = CAR_VERSOES[tabela]
    
    q = f"""
    (
        SELECT DISTINCT ON ({CAR_ID})
            {i} AS versao,
            DATE '{data_ref}' AS data_ref,
            {CAR_ID},
            split_part({CAR_ID}, '-', 2) AS cod_mun,
            municipio,
            {CAR_GEOM} AS geom
        FROM {CAR_SCHEMA}.{tabela}
        WHERE {CAR_ID} = '{COD_IMOVEL}'
          AND {CAR_STATUS} <> 'CA'
          AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
        ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
    )
    """
    queries.append(q)

query_compara_geometria = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
),
comparacao AS (
    SELECT
        *,
        LAG(geom) OVER (ORDER BY versao) AS geom_anterior,
        LAG(versao) OVER (ORDER BY versao) AS versao_anterior
    FROM car_series
)
SELECT
    versao_anterior,
    versao,
    CASE 
        WHEN NOT ST_Equals(geom_anterior, geom) THEN TRUE
        ELSE FALSE
    END AS alteracao_geometria
FROM comparacao
WHERE geom_anterior IS NOT NULL;
"""

# ============================================================================
# Exibição da query
# ============================================================================

html_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em; margin-bottom: 15px;">
<b>Consulta SQL - Verificacao de alteracao de geometria (ST_Equals):</b><br>
<code>
{query_compara_geometria}
</code>
</div>
"""

display(HTML(html_query))
```

#### Execução das querys

```python
# ============================================================================
# Executa a consulta e exibe os resultados
# ============================================================================

df_compara_geometria = pd.read_sql(text(query_compara_geometria), con=engine)

# ============================================================================
# Balão de resumo (baseado nos resultados)
# ============================================================================

alteracoes = df_compara_geometria[df_compara_geometria['alteracao_geometria'] == True]
num_alteracoes = len(alteracoes)

if num_alteracoes == 0:
    balao_resumo = """
    <div style="background-color:#e6f4ea; padding:15px; border-radius:8px; border-left: 5px solid #4caf50; margin-bottom: 15px;">
    <b>Resumo da verificação:</b><br>
    Não houve alterações na geometria entre as versões analisadas.<br>
    </div>
    """
else:
    balao_resumo = f"""
    <div style="background-color:#fff3cd; padding:15px; border-radius:8px; border-left: 5px solid #ff9800; margin-bottom: 15px;">
    <b>Resumo da verificação:</b><br>
    Foram detectadas <b>{num_alteracoes}</b> alteração(ões) na geometria entre as versões.<br>
    As alterações ocorreram nos seguintes pares:
    <ul>
    """
    for _, row in alteracoes.iterrows():
        balao_resumo += f"<li>Versão {int(row['versao_anterior'])} → Versão {int(row['versao'])}</li>"
    balao_resumo += """
    </div>
    """

# ============================================================================
# Exibe o balão de resumo + a tabela de resultados
# ============================================================================

display(HTML(balao_resumo))

html_resultado = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin-top: 15px;">
<b>Resultado detalhado - Alteração de geometria entre versões</b><br>
<br>
{df_compara_geometria.to_html(index=False)}
<br><br>
<b>Legenda:</b><br>
- <b>TRUE</b>: houve alteração na geometria entre as versões<br>
- <b>FALSE</b>: as geometrias são exatamente iguais 
</div>
"""

display(HTML(html_resultado))
```

### 4.2.3: Calcular a área e porte do imóvel

Nesta etapa, vamos calcular para cada versão do CAR:

- **Área do imóvel** (em hectares), utilizando a projeção Albers (SRID 97823)
- **Porte do imóvel** (Pequeno, Médio, Grande), com base no módulo fiscal do município

O porte é definido pela relação entre a área do imóvel e o módulo fiscal do município:

| Porte | Quantidade de módulos fiscais |
|-------|-------------------------------|
| Pequeno | < 4 |
| Médio | 4 a 15 |
| Grande | > 15 |

O número de módulos fiscais é obtido pela razão entre a área do imóvel e o valor do módulo fiscal do município. O módulo fiscal do município é obtido a partir da tabela de módulos fiscais do INCRA, utilizando o código do município extraído do `cod_imovel`.

#### Cálculo de área e porte

```python
from IPython.display import HTML, display
from sqlalchemy import text

# ============================================================================
# Query para calcular área, módulo fiscal e porte do imóvel
# ============================================================================

queries = []

for i, tabela in enumerate(CAR_TABLES, start=1):
    data_ref = CAR_VERSOES[tabela]
    
    q = f"""
    (
        SELECT DISTINCT ON ({CAR_ID})
            {i} AS versao,
            DATE '{data_ref}' AS data_ref,
            {CAR_ID},
            split_part({CAR_ID}, '-', 2) AS cod_mun,
            municipio,
            {CAR_GEOM} AS geom
        FROM {CAR_SCHEMA}.{tabela}
        WHERE {CAR_ID} = '{COD_IMOVEL}'
          AND {CAR_STATUS} <> 'CA'
          AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
        ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
    )
    """
    queries.append(q)

query_tabela_base = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
)
SELECT 
    a.versao,
    a.data_ref,
    a.{CAR_ID},
    a.cod_mun,
    a.municipio,
    ST_Area(ST_Transform(a.geom, 97823)) / 10000 AS area_imovel_ha,
    b.modulofiscal_ha,
    (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha AS num_mf,
    CASE
        WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha <= 4 THEN 'pequeno'
        WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha > 15 THEN 'grande'
        ELSE 'medio'
    END AS tamanho_imovel,
    a.geom
FROM car_series a 
LEFT JOIN {MODFISCAL_SCHEMA}.{MODFISCAL_TABLE} b ON b.{MODFISCAL_CDMUN}::text = a.cod_mun
ORDER BY versao;
"""

# ============================================================================
# Exibição da query em um balão bonito
# ============================================================================

html_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em; margin-bottom: 15px;">
<b>Consulta SQL - Calculo de area, modulo fiscal e porte do imovel:</b><br>
<code>
{query_tabela_base}
</code>
</div>
"""

display(HTML(html_query))
```

#### Executa a query

```python
# ============================================================================
# Executa a consulta
# ============================================================================

df_tabela_base = pd.read_sql(text(query_tabela_base), con=engine)

# ============================================================================
# Formata a coluna geom para mostrar apenas os primeiros caracteres
# ============================================================================

def formatar_geom(geom_str):
    if isinstance(geom_str, str) and len(geom_str) > 50:
        resumo = geom_str[:50] + "..."
        return f'<details><summary style="cursor:pointer; color:#2196f3;">{resumo}</summary><code style="word-break:break-all;">{geom_str}</code></details>'
    return geom_str

df_display = df_tabela_base.copy()
if 'geom' in df_display.columns:
    df_display['geom'] = df_display['geom'].apply(formatar_geom)

# ============================================================================
# Exibe o resultado em um balão bonito
# ============================================================================

html_resultado = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin-top: 15px;">
<b>Resultado - Area e porte do imovel por versao</b><br>
<br>
{df_display.to_html(escape=False, index=False)}
</div>
"""

display(HTML(html_resultado))
```

### 4.2.4: Comparação entre versões do cadastro

Nesta etapa, comparamos versões consecutivas do imóvel, com o objetivo de identificar as mudanças ocorridas ao longo do tempo.

São avaliados os seguintes aspectos:

- **Alteração geométrica (`alteracao_geometria`)**: indica se houve mudança na geometria do imóvel (métrica já computada anteriormente e reapresentada aqui);
- **Mudança de porte (`houve_mudanca_porte`)**: verifica se houve alteração na classificação do imóvel (pequeno, médio ou grande);
- **Variação de área (`perc_var_area`)**: quantifica a diferença percentual entre as áreas em versões consecutivas;
- **Percentual de sobreposição entre versões (`perc_sobreposicao`)**: mede o percentual de interseção entre as geometrias da versão anterior e da versão atual.

#### Comparando as versões

```python
# ============================================================================
# Query para comparar versões do cadastro 
# ============================================================================

queries = []

for i, tabela in enumerate(CAR_TABLES, start=1):
    data_ref = CAR_VERSOES[tabela]
    
    q = f"""
    (
        SELECT DISTINCT ON ({CAR_ID})
            {i} AS versao,
            DATE '{data_ref}' AS data_ref,
            {CAR_ID},
            split_part({CAR_ID}, '-', 2) AS cod_mun,
            municipio,
            {CAR_GEOM} AS geom
        FROM {CAR_SCHEMA}.{tabela}
        WHERE {CAR_ID} = '{COD_IMOVEL}'
          AND {CAR_STATUS} <> 'CA'
          AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
        ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
    )
    """
    queries.append(q)

query_tabela_comparacao = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
), 
tabela_base AS (
    SELECT 
        a.versao,
        a.data_ref,
        a.{CAR_ID},
        a.cod_mun,
        a.municipio,
        ST_Area(ST_Transform(a.geom, 97823)) / 10000 AS area_imovel_ha,
        b.modulofiscal_ha,
        (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha AS num_mf,
        CASE
            WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha <= 4 THEN 'pequeno'
            WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha > 15 THEN 'grande'
            ELSE 'medio'
        END AS tamanho_imovel,
        a.geom
    FROM car_series a 
    LEFT JOIN {MODFISCAL_SCHEMA}.{MODFISCAL_TABLE} b ON b.{MODFISCAL_CDMUN}::text = a.cod_mun
    ORDER BY versao
),
comparacao AS (
    SELECT
        *,
        LAG(geom) OVER (ORDER BY versao) AS geom_anterior,
        LAG(versao) OVER (ORDER BY versao) AS versao_anterior,
        LAG(tamanho_imovel) OVER (ORDER BY versao) AS tamanho_anterior
    FROM tabela_base
)
SELECT
    versao_anterior,
    versao,
    CASE 
        WHEN NOT ST_Equals(geom_anterior, geom) THEN TRUE
        ELSE FALSE
    END AS alteracao_geometria,
    CASE
        WHEN tamanho_anterior <> tamanho_imovel THEN TRUE
        ELSE FALSE
    END AS houve_mudanca_porte,
    ((ST_Area(ST_Transform(geom, 97823)) / NULLIF(ST_Area(ST_Transform(geom_anterior, 97823)), 0)) - 1.0) * 100 AS perc_var_area,
    ST_Area(ST_Transform(ST_Intersection(geom_anterior, geom), 97823)) / 10000 AS area_intersect_ha,
    (ST_Area(ST_Transform(ST_Intersection(geom_anterior, geom), 97823)) / NULLIF(ST_Area(ST_Transform(geom_anterior, 97823)), 0)) * 100 AS perc_sobreposicao
FROM comparacao
WHERE geom_anterior IS NOT NULL;
"""

# ============================================================================
# Exibição da query
# ============================================================================

html_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em; margin-bottom: 15px;">
<b>Consulta SQL - Comparacao entre versoes do cadastro:</b><br>
<code>
{query_tabela_comparacao}
</code>
</div>
"""

display(HTML(html_query))
```

#### Executa

```python
# ============================================================================
# Executa a consulta
# ============================================================================

df_comparacao = pd.read_sql(text(query_tabela_comparacao), con=engine)

# ============================================================================
# Arredonda as colunas numéricas no Python (já que o PostgreSQL não tem ROUND com 2 args)
# ============================================================================

df_comparacao['perc_var_area'] = df_comparacao['perc_var_area'].apply(lambda x: round(x, 2) if pd.notna(x) else x)
df_comparacao['area_intersect_ha'] = df_comparacao['area_intersect_ha'].apply(lambda x: round(x, 2) if pd.notna(x) else x)
df_comparacao['perc_sobreposicao'] = df_comparacao['perc_sobreposicao'].apply(lambda x: round(x, 2) if pd.notna(x) else x)

# ============================================================================
# Exibe o resultado
# ============================================================================

html_resultado = f"""
<div style="background-color:#e8f4fd; padding:15px; border-radius:8px; border-left: 5px solid #2196f3; margin-top: 15px;">
<b>Resultado - Comparação entre versões do cadastro</b><br>
<br>
{df_comparacao.to_html(index=False)}
<br><br>
<b>Legenda:</b><br>
- <b>alteracao_geometria</b>: TRUE se houve mudança na geometria (ST_Equals)<br>
- <b>houve_mudanca_porte</b>: TRUE se o porte do imóvel mudou<br>
- <b>perc_var_area</b>: variação percentual da área em relação à versão anterior<br>
- <b>area_intersect_ha</b>: área de intersecção entre versões (hectares)<br>
- <b>perc_sobreposicao</b>: percentual de sobreposição de área entre as versões
</div>
"""

display(HTML(html_resultado))
```

```python
from IPython.display import HTML, display
from sqlalchemy import text

# ============================================================================
# Query para calcular área, módulo fiscal e porte do imóvel
# ============================================================================

queries = []

for i, tabela in enumerate(CAR_TABLES, start=1):
    data_ref = CAR_VERSOES[tabela]
    
    q = f"""
    (
        SELECT DISTINCT ON ({CAR_ID})
            {i} AS versao,
            DATE '{data_ref}' AS data_ref,
            {CAR_ID},
            split_part({CAR_ID}, '-', 2) AS cod_mun,
            municipio,
            {CAR_GEOM} AS geom
        FROM {CAR_SCHEMA}.{tabela}
        WHERE {CAR_ID} = '{COD_IMOVEL}'
          AND {CAR_STATUS} <> 'CA'
          AND {CAR_CONDICAO} NOT ILIKE '%cancelado%'
        ORDER BY {CAR_ID}, COALESCE({CAR_DATA_ATUALIZACAO}, {CAR_DATA_CRIACAO}) DESC
    )
    """
    queries.append(q)

query_tabela_base_final = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
)
SELECT 
    a.versao,
    a.data_ref,
    a.{CAR_ID},
    a.cod_mun,
    a.municipio,
    ST_Area(ST_Transform(a.geom, 97823)) / 10000 AS area_imovel_ha,
    b.modulofiscal_ha,
    (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha AS num_mf,
    CASE
        WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha <= 4 THEN 'pequeno'
        WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha > 15 THEN 'grande'
        ELSE 'medio'
    END AS tamanho_imovel,
    a.geom
FROM car_series a 
LEFT JOIN {MODFISCAL_SCHEMA}.{MODFISCAL_TABLE} b ON b.{MODFISCAL_CDMUN}::text = a.cod_mun
ORDER BY versao;
"""


query_tabela_comparacao_final = f"""
WITH car_series AS (
    {' UNION ALL '.join(queries)}
), 
tabela_base AS (
    SELECT 
        a.versao,
        a.data_ref,
        a.{CAR_ID},
        a.cod_mun,
        a.municipio,
        ST_Area(ST_Transform(a.geom, 97823)) / 10000 AS area_imovel_ha,
        b.modulofiscal_ha,
        (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha AS num_mf,
        CASE
            WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha <= 4 THEN 'pequeno'
            WHEN (ST_Area(ST_Transform(a.geom, 97823)) / 10000) / b.modulofiscal_ha > 15 THEN 'grande'
            ELSE 'medio'
        END AS tamanho_imovel,
        a.geom
    FROM car_series a 
    LEFT JOIN {MODFISCAL_SCHEMA}.{MODFISCAL_TABLE} b ON b.{MODFISCAL_CDMUN}::text = a.cod_mun
    ORDER BY versao
),
comparacao AS (
    SELECT
        *,
        LAG(geom) OVER (ORDER BY versao) AS geom_anterior,
        LAG(versao) OVER (ORDER BY versao) AS versao_anterior,
        LAG(tamanho_imovel) OVER (ORDER BY versao) AS tamanho_anterior
    FROM tabela_base
)
SELECT
    versao_anterior,
    versao,
    CASE 
        WHEN NOT ST_Equals(geom_anterior, geom) THEN TRUE
        ELSE FALSE
    END AS alteracao_geometria,
    CASE
        WHEN tamanho_anterior <> tamanho_imovel THEN TRUE
        ELSE FALSE
    END AS houve_mudanca_porte,
    ((ST_Area(ST_Transform(geom, 97823)) / NULLIF(ST_Area(ST_Transform(geom_anterior, 97823)), 0)) - 1.0) * 100 AS perc_var_area,
    ST_Area(ST_Transform(ST_Intersection(geom_anterior, geom), 97823)) / 10000 AS area_intersect_ha,
    (ST_Area(ST_Transform(ST_Intersection(geom_anterior, geom), 97823)) / NULLIF(ST_Area(ST_Transform(geom_anterior, 97823)), 0)) * 100 AS perc_sobreposicao
FROM comparacao
WHERE geom_anterior IS NOT NULL;
"""

# ============================================================================
# Exibição da query em um balão bonito
# ============================================================================

html_query = f"""
<div style="background-color:#f5f5f5; padding:15px; border-radius:8px; border-left: 5px solid #2c6e9e; font-family: monospace; font-size: 0.9em; margin-bottom: 15px;">
<b>Consulta SQL - Calculo de area, modulo fiscal e porte do imovel:</b><br>
<code>
{query_tabela_base_final}
</code>
<br><br><br>
<b>Consulta SQL - Comparacao entre versoes do cadastro:</b><br>
<code>
{query_tabela_comparacao_final}
</code>
</div>
"""


display(HTML(html_query))
```

##### Executa

```python
# ============================================================================
# Executa a consulta
# ============================================================================

df_tabela_base_final = pd.read_sql(text(query_tabela_base_final), con=engine)

df_comparacao_final = pd.read_sql(text(query_tabela_comparacao_final), con=engine)
```

#### Visualização dos imóveis

```python
# ============================================================================
# Visualização final 
# ============================================================================

if 'df_tabela_base_final' not in locals() or df_tabela_base_final.empty:
    print("X Execute a primeira célula com a query da tabela base da 4.2.5.")
else:
    # -------------------------------------------------------------------------
    # Define cores para cada versão (mesma lógica do mapa anterior)
    # -------------------------------------------------------------------------
    versoes_unicas = sorted(df_tabela_base_final['versao'].unique())
    total_versoes = len(versoes_unicas)
    
    def get_cor_versao(versao_num, total_versoes):
        # Borda: do cinza claro ao cinza escuro
        intensidade = 0.3 + (versao_num - 1) * 0.5 / max(1, total_versoes - 1)
        intensidade = min(intensidade, 0.8)
        cor_borda = f"rgb({int(128 + intensidade * 127)}, {int(128 + intensidade * 127)}, {int(128 + intensidade * 127)})"
        return cor_borda
    
    # -------------------------------------------------------------------------
    # Função para converter WKB para geometria
    # -------------------------------------------------------------------------
    def wkb_to_geom(wkb_str):
        if isinstance(wkb_str, str):
            geom_bytes = binascii.unhexlify(wkb_str)
            return wkb.loads(geom_bytes)
        return wkb_str
    
    # -------------------------------------------------------------------------
    # Prepara o mapa base
    # -------------------------------------------------------------------------
    primeira_linha = df_tabela_base_final.iloc[0]
    primeira_geom = wkb_to_geom(primeira_linha['geom'])
    bounds = primeira_geom.bounds
    map_center = [(bounds[1] + bounds[3]) / 2, (bounds[0] + bounds[2]) / 2]
    
    m_final = folium.Map(
        location=map_center,
        zoom_start=10,
        control_scale=True,
        tiles='Esri.WorldImagery',
        width='80%',
        height=500
    )
    m_final.fit_bounds([[bounds[1], bounds[0]], [bounds[3], bounds[2]]])
    
    # -------------------------------------------------------------------------
    # Adiciona cada versão ao mapa
    # -------------------------------------------------------------------------
    for idx, row in df_tabela_base_final.iterrows():
        versao_num = int(row['versao'])
        data_ref = row['data_ref']
        cod_imovel = row['cod_imovel']
        municipio = row['municipio']
        area_ha = row['area_imovel_ha']
        mod_fiscal_ha = row['modulofiscal_ha']
        num_mf = row['num_mf']
        tamanho = row['tamanho_imovel']
        
        # Converte a geometria
        geom = wkb_to_geom(row['geom'])
        
        # Prepara geometria para o Folium
        gdf_temp = gpd.GeoDataFrame({'geometry': [geom]}, crs=4674)
        gdf_json_safe = gdf_temp.copy()
        
        # Define a cor da borda baseada na versão
        cor_borda = get_cor_versao(versao_num, total_versoes)
        
        # Tooltip com todas as informações
        tooltip_text = f"""
        <div style="font-family: Arial, sans-serif; min-width: 280px;">
            <b>Versao {versao_num} ({data_ref})</b><br>
            <hr style="margin: 5px 0;">
            <b>Codigo:</b> {cod_imovel}<br>
            <b>Municipio:</b> {municipio}<br>
            <b>Area:</b> {float(area_ha):.2f} ha<br>
            <b>Modulo fiscal (INCRA):</b> {float(mod_fiscal_ha):.2f} ha<br>
            <b>Num. modulos fiscais:</b> {float(num_mf):.2f}<br>
            <b>Porte:</b> {tamanho}
        </div>
        """
        
        folium.GeoJson(
            gdf_json_safe,
            name=f"Versao {versao_num} ({data_ref})",
            style_function=lambda x, cor=cor_borda: {
                'color': cor,
                'weight': 2.5,
                'fillColor': 'white',
                'fillOpacity': 0.35
            },
            tooltip=folium.Tooltip(tooltip_text)
        ).add_to(m_final)
    
    # -------------------------------------------------------------------------
    # Adiciona controle de camadas e exibe
    # -------------------------------------------------------------------------
    folium.LayerControl().add_to(m_final)
    
    display(HTML("""
    <div style="background-color:#e8f4fd; padding:12px; border-radius:8px; border-left: 5px solid #2196f3; margin-bottom: 15px;">
    <b>Mapa final - Evolucao do imovel ao longo das versoes</b><br>
    <small>Cada versao pode ser ligada/desligada no controle de camadas (canto superior direito)</small>
    </div>
    """))
    
    display(m_final)
    
    # Libera memória
    del m_final
    gc.collect()
```
